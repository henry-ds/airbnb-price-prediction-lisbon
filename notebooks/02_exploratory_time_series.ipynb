# Exploratory Time Series Analysis of Airbnb Demand

This notebook explores the temporal characteristics of Airbnb demand in Lisbon
using monthly aggregated review data. The analysis focuses on identifying
trend, seasonality, and irregular components to inform subsequent forecasting
models.


## Objective of Analysis

The objective of this exploratory analysis is to examine the structure of the
time series and assess the presence of long-term trends and seasonal patterns.
Understanding these characteristics is essential for selecting appropriate
forecasting models and parameters.


```python
import pandas as pd
import os
import matplotlib.pyplot as plt
# Import seasonal decomposition function from statsmodels
from statsmodels.tsa.seasonal import seasonal_decompose

# Load time series data from CSV file
ts_df = pd.read_csv("../data/monthly_demand.csv")

# Convert 'date' column to datetime format
ts_df['date'] = pd.to_datetime(ts_df['date'])

# Set 'date' column as index for time series analysis
ts_df.set_index('date', inplace=True)
ts_df.head()
```

## Seasonal Decomposition

Seasonal decomposition is applied to separate the observed time series into
three components: trend, seasonality, and residuals. This process enables a
clearer interpretation of the underlying demand dynamics.

An additive decomposition model is employed, as seasonal fluctuations appear to
be relatively constant in magnitude across time. A seasonal period of 12 is
used to reflect monthly seasonality.


```python


# Perform seasonal decomposition on the 'reviews_count' column
# Using additive model with period=12 (monthly seasonality)
decomposition = seasonal_decompose(
    ts_df['reviews_count'],
    model='additive',
    period=12
)

# Plot the decomposition components (trend, seasonal, residual)
decomposition.plot()
plt.show()
```

## Interpretation of Decomposition Results

The decomposition reveals a clear seasonal pattern in Airbnb demand, with
higher activity observed during the summer months. This pattern is consistent
with Lisbons tourism-driven short-term rental market.

The trend component reflects long-term changes in demand, while the residual
component captures short-term irregular variations that are not explained by
trend or seasonality alone.
