This notebook explores the cleaned Airbnb listings dataset to identify patterns,
relationships, and potential drivers of rental prices in Lisbon.


## Imports

```python
# Import pandas library for data manipulation and analysis
import pandas as pd
# Import matplotlib.pyplot for creating static visualizations
import matplotlib.pyplot as plt
# Import seaborn
import seaborn as sns
```

## Load Clean Data

```python
# Load the cleaned listings data from CSV file into a pandas DataFrame
df = pd.read_csv("../data/processed/clean_listings.csv")
```

## Price Distribution

```python
# Create a figure with specified dimensions (8x4 inches)
plt.figure(figsize=(8,4))
# Create a histogram of the 'price' column from the dataframe with 50 bins
sns.histplot(df["price"], bins=50)
# Add a title to the plot
plt.title("Distribution of Airbnb Prices in Lisbon")
# Display the plot
plt.show()
```

## Price by Room Type


```python
# Create a figure with specified size (8 inches width, 4 inches height)
plt.figure(figsize=(8,4))
# Create a boxplot showing price distribution across different room types
sns.boxplot(x="room_type", y="price", data=df)
# Add a title to the plot
plt.title("Price by Room Type")
# Display the plot
plt.show()
```

## Correlation Heatmap

```python
# Create a figure with specified dimensions (6x4 inches)
plt.figure(figsize=(6,4))
# Generate a heatmap showing correlation between numerical features
sns.heatmap(df.corr(numeric_only=True), annot=True)
# Add a title to the heatmap
plt.title("Correlation Between Numerical Features")
# Display the plot
plt.show()
```

## Save Visual

```python
# Create a figure with specified dimensions (8x4 inches)
plt.figure(figsize=(8,4))
# Create a histogram of the 'price' column with 50 bins
sns.histplot(df["price"], bins=50)
# Add a title to the plot
plt.title("Distribution of Airbnb Prices in Lisbon")
# Adjust layout to prevent overlapping elements
plt.tight_layout()
# Save the figure to the specified path
plt.savefig("../visuals/price_distribution.png")
# Display the plot
plt.show()
```

## Key Insights

- Airbnb prices in Lisbon are right-skewed, with most listings below â‚¬200 per night.
- Entire homes command significantly higher prices than private rooms.
- Location and availability show meaningful relationships with price.
